0/1 Knapsack Problem

* Analyze three different techniques to solve knapsacks problem
1. Recursion
2. Dynamic Programming
3. Memoization

```python
#sample code for knapsack problem using recursion
def rec_knapSack(w, wt, val, n):

  #base case
  #defined as nth item is empty;
  #or the capacity w is 0
  if n == 0 or w == 0:
    return 0

  #if weight of the nth item is more than
  #the capacity W, then this item cannot be included
  #as part of the optimal solution
  if(wt[n-1] > w):
    return rec_knapSack(w, wt, val, n-1)

  #return the maximum of the two cases:
  # (1) include the nth item
  # (2) don't include the nth item
  else:
    return max(
        val[n-1] + rec_knapSack(
            w-wt[n-1], wt, val, n-1),
            rec_knapSack(w, wt, val, n-1)
    )
```

```python
#To test:
val = [60, 100, 120] #values for the items
wt = [10, 20, 30] #weight of the items
w = 50 #knapsack weight capacity
n = len(val) #number of items

rec_knapSack(w, wt, val, n)
```

```python
#Dynamic Programming for the Knapsack Problem
def DP_knapSack(w, wt, val, n):
  #create the table
  table = [[0 for x in range(w+1)] for x in range (n+1)]

  #populate the table in a bottom-up approach
  for i in range(n+1):
    for w in range(w+1):
      if i == 0 or w == 0:
        table[i][w] = 0
      elif wt[i-1] <= w:
        table[i][w] = max(val[i-1] + table[i-1][w-wt[i-1]],
                          table[i-1][w])
  return table[n][w]
```

```python
#To test:
val = [60, 100, 120]
wt = [10, 20, 30]
w = 50
n = len(val)

DP_knapSack(w, wt, val, n)
```

```python
#Sample for top-down DP approach (memoization)
#initialize the list of items
val = [60, 100, 120]
wt = [10, 20, 30]
w = 50
n = len(val)

#initialize the container for the values that have to be stored
#values are initialized to -1
calc =[[-1 for i in range(w+1)] for j in range(n+1)]


def mem_knapSack(wt, val, w, n):
  #base conditions
  if n == 0 or w == 0:
    return 0
  if calc[n][w] != -1:
    return calc[n][w]

  #compute for the other cases
  if wt[n-1] <= w:
    calc[n][w] = max(val[n-1] + mem_knapSack(wt, val, w-wt[n-1], n-1),
                     mem_knapSack(wt, val, w, n-1))
    return calc[n][w]
  elif wt[n-1] > w:
    calc[n][w] = mem_knapSack(wt, val, w, n-1)
    return calc[n][w]

mem_knapSack(wt, val, w, n)
```

**Code Analysis**

#Recursion
This code for knapsack problem using recursion, is solving the knapsack problem using recursion. The idea is that we have a bag with a certain weight capacity and a set of items, each with its own weight and value. The function rec_knapSack takes in the capacity of the bag (w), the list of weights (wt), the list of values (val), and the number of items (n). If there are no items left or the capacity is zero, the function just returns 0 because nothing can be added. If the current item is too heavy to fit in the remaining capacity, the function skips it and checks the rest. Otherwise, the function has two choices: include the current item or exclude it. If it includes the item, it adds the item’s value and reduces the capacity by the item’s weight, then solves the smaller problem with the remaining items. If it excludes the item, it just solves the smaller problem without changing the capacity. The function then returns whichever choice gives the bigger value. When tested with values `[60, 100, 120]` and weights `[10, 20, 30]` with a capacity of `50`, the best combination is the second and third items, giving a total value of `220`. This recursive approach works, but it’s slow for bigger inputs because it repeats a lot of calculations. That’s why dynamic programming is usually used to make it faster.

#Dynamic Programming
This code solves the knapsack problem using dynamic programming in a bottom‑up way. It builds a 2D table where each entry represents the best value achievable with a certain number of items and capacity. The first row and column are set to zero since no items or no capacity means no value. For each item, if its weight is less than or equal to the current capacity, the algorithm compares two options: including the item (its value plus the best value for the remaining capacity) or excluding it (keeping the previous best). The maximum of these is stored in the table. The final answer is found in the bottom‑right cell, which for the test case `[60, 100, 120]` with weights `[10, 20, 30]` and capacity `50` gives `220`. This approach is efficient because it avoids repeated calculations that occur in recursion.

#Memoization
This code is solving the knapsack problem using memoization, which is a top‑down dynamic programming approach. It works like recursion but stores results in a table called `calc` so the same subproblems don’t get solved over and over. The function checks the base case first: if there are no items or no capacity, the answer is 0. If the result for the current state is already in `calc`, it just returns that. Otherwise, it decides whether the current item can fit in the knapsack. If it can, it compares the value of including the item versus excluding it, and takes the larger one. If it’s too heavy, it skips the item. The best result is saved in `calc` and returned. With the test values `[60, 100, 120]` and weights `[10, 20, 30]` at capacity 50, the maximum value is 220. This method is faster than plain recursion because it avoids repeating the same calculations.

```python
#Recursion

# Recursion with weight and volume constraints
# W = remaining weight capacity
# V = remaining volume capacity
# wt = list of item weights
# vol = list of item volumes
# val = list of item values
# n = number of items

def rec_knapSack(W, V, val, wt, vol, n):
    if n == 0 or W == 0 or V == 0:
        return 0

    pick = 0

    if wt[n-1] <= W and vol[n-1] <= V:
        pick = val[n-1] + rec_knapSack(W - wt[n-1], V - vol[n-1], val, wt, vol, n-1)

    notPick = rec_knapSack(W, V, val, wt, vol, n-1)

    return max(pick, notPick)


def knapsack(W, V, val, wt, vol):
    n = len(val)
    return rec_knapSack(W, V, val, wt, vol, n)


if __name__ == "__main__":
    val = [60, 100, 120]
    wt = [10, 20, 30]
    vol = [2, 3, 4]
    W = 50
    V = 6

    print(knapsack(W, V, val, wt, vol))
```

```python
# Dynamic Programming

def DP_knapSack_3D(max_budget, max_kg, costs, weights, vals, n):
    table = [[[0 for _ in range(max_kg + 1)]
               for _ in range(max_budget + 1)]
               for _ in range(n + 1)]

    for i in range(n + 1):
        for b in range(max_budget + 1):
            for k in range(max_kg + 1):
                if i == 0 or b == 0 or k == 0:
                    table[i][b][k] = 0

                elif costs[i-1] <= b and weights[i-1] <= k:
                    include = vals[i-1] + table[i-1][b - costs[i-1]][k - weights[i-1]]
                    skip = table[i-1][b][k]
                    table[i][b][k] = max(include, skip)
                else:
                    table[i][b][k] = table[i-1][b][k]

    return table[n][max_budget][max_kg]

# To test:
nutrition = [100, 40, 30, 10]
costs = [280, 150, 100, 80]
kgs = [5, 2, 3, 4]
my_budget = 500
my_limit_kg = 10
n_items = len(nutrition)

result = DP_knapSack_3D(my_budget, my_limit_kg, costs, kgs, nutrition, n_items)

print(f"Budget: ₱{my_budget} | Weight Limit: {my_limit_kg}kg")
print(f"Max Nutrition: {result}")
```

```python
#Memoization Programming

# 2 Constraints (Weight & Volume)
val = [60, 100, 120]
wt = [10, 20, 30]
vol = [5, 15, 10]
w_limit = 50
v_limit = 20
n = len(val)

calc = [[[-1 for v in range(v_limit + 1)]
          for w in range(w_limit + 1)]
          for i in range(n + 1)]

def mem_knapSack_3D(wt, vol, val, w, v, n):
    if n == 0 or w == 0 or v == 0:
        return 0

    if calc[n][w][v] != -1:
        return calc[n][w][v]

    if wt[n-1] <= w and vol[n-1] <= v:
        take_it = val[n-1] + mem_knapSack_3D(wt, vol, val, w - wt[n-1], v - vol[n-1], n - 1)
        skip_it = mem_knapSack_3D(wt, vol, val, w, v, n - 1)

        calc[n][w][v] = max(take_it, skip_it)
    else:
        calc[n][w][v] = mem_knapSack_3D(wt, vol, val, w, v, n - 1)

    return calc[n][w][v]

# Test out:
result = mem_knapSack_3D(wt, vol, val, w_limit, v_limit, n)
print(f"Max Value with Weight & Volume constraints: {result}")
```

Task 2: Create a sample program that find the nth number of Fibonacci Series using Dynamic Programming

```python
def get_nth_fibonacci(n):
    if n <= 1:
        return n

    history = [0] * (n + 1)

    history[0] = 0
    history[1] = 1

    for i in range(2, n + 1):
        history[i] = history[i-1] + history[i-2]

    return history[n]

# Run Test
n_target = 10
result = get_nth_fibonacci(n_target)

print(f"The {n_target}th Fibonacci number is: {result}")
```

```python
#Recursive programming

def grocery_optimized_ph(budget, costs, nutrition):
    n = len(nutrition)
    dp = [[0 for _ in range(budget + 1)] for _ in range(n + 1)]

    for i in range(1, n + 1):
        for w in range(1, budget + 1):
            if costs[i-1] <= w:
                dp[i][w] = max(nutrition[i-1] + dp[i-1][w - costs[i-1]],
                               dp[i-1][w])
            else:
                dp[i][w] = dp[i-1][w]
    return dp[n][budget]

# Items
item_names = ["5kg Rice", "Chicken", "Eggs", "Sardines", "Tilapia", "Veggies", "Oil"]
nutrition_points = [120, 90, 60, 45, 75, 50, 10]
cost_in_pesos = [280, 220, 115, 150, 180, 90, 75]
total_budget_php = 1000

max_nutrition = grocery_optimized_ph(total_budget_php, cost_in_pesos, nutrition_points)

print(f"--- Grocery Run Result ---")
print(f"Total Budget: ₱{total_budget_php}")
print(f"Highest Nutrition Score: {max_nutrition} points")
```

```python
# Dynamic programming

def solve_survival_kit(budget, costs, satisfaction):
    n = len(satisfaction)

    dp = [[0 for _ in range(budget + 1)] for _ in range(n + 1)]

    for i in range(1, n + 1):
        for w in range(1, budget + 1):
            if costs[i-1] <= w:
                dp[i][w] = max(satisfaction[i-1] + dp[i-1][w - costs[i-1]],
                               dp[i-1][w])
            else:

                dp[i][w] = dp[i-1][w]

    return dp[n][budget]

# Petsa de Peligro days
items = ["Noodles", "Rice", "Corned Beef", "Coffee", "Tuna", "Bread", "Siomai"]
costs = [60, 55, 85, 110, 45, 70, 150]
fullness = [30, 80, 60, 40, 50, 45, 70]
my_budget = 500

max_fullness = solve_survival_kit(my_budget, costs, fullness)

print(f"Remaining Budget: ₱{my_budget}")
print(f"Max Fullness Score: {max_fullness} points")
```
