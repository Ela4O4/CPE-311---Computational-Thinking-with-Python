4. Isolate the fare_amount, tip_amount, tolls_amount, and total_amount for the longest trip by distance (trip_distance).

```python
#To find the row where the trip_distance is at its maximum
longest_trip = taxi_df[taxi_df['trip_distance'] == taxi_df['trip_distance'].max()]

#To Select only the four requested financial columns for that row
result = longest_trip[['fare_amount', 'tip_amount', 'tolls_amount', 'total_amount']]

#To Display the result
result
```
