```python
import matplotlib.pyplot as plt
import numpy as np

# Data
bias = 1
weight = 3
x = 0

# Calculate prediction: y = wx + b
y_val = (weight * x) + bias

# Plotting
x_line = np.linspace(-3, 3, 100)
y_line = (weight * x_line) + bias

# Plot
plt.figure(figsize=(6, 4))
plt.plot(x_line, y_line, color='blue', label='Regression Line')
plt.scatter(x, y_val, color='red', s=100, label=f'Result (x=0, y={y_val})')
plt.title("Part 1: Simple Linear Regression")
plt.xlabel("Input (x)")
plt.ylabel("Output (y)")
plt.legend()
plt.grid(True)
plt.show()

print(f"Prediction for Part 1: {y_val}")
```

```python
b = 1
w = np.array([3, -2, 4])
x_in = np.array([[3, -1, 2], [10, 0, -10]])

# This calculates the Raw Scores (z) used in the plot
z = np.dot(x_in, w) + b


# This function squashes any number to be between 0 and 1
def sigmoid(z_val):
    return 1 / (1 + np.exp(-z_val))

# Turns the raw scores into probabilities (0% to 100%)
p = sigmoid(z)

# Draw the S-curve to show how the scores are mapped
z_space = np.linspace(-15, 25, 100)
plt.figure(figsize=(6, 4))
plt.plot(z_space, sigmoid(z_space), color='purple', label='S-Curve')
plt.scatter(z, p, color='orange', zorder=5, label='Our Examples')

# Add the 0.5 cutoff line
plt.axhline(0.5, color='black', linestyle='--', label='50% Cutoff')

plt.title("Turning Scores into Classes")
plt.xlabel("Raw Score")
plt.ylabel("Probability")
plt.legend()
plt.show()

# Shows the raw score (z) and the final probability (p)
print(f"Example 1: Raw Score (z) = {z[0]}, Prob = {p[0]:.2f}")
print(f"Example 2: Raw Score (z) = {z[1]}, Prob = {p[1]:.2f}")

# Final result
# If probability is > 0.5, it's Class 1. If lower, it's Class 0.
print(f"Example 1: Prob is {p[0]:.2f} -> Result: Class 1")
print(f"Example 2: Prob is {p[1]:.2f} -> Result: Class 0")
```
