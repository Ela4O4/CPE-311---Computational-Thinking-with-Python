```python
import random
random.seed(0)
salaries = [round(random.random()*1000000, -3) for _ in range(100)]
```

Using the data generated above, calculate the following statistics without importing anything from the statistics module in the
standard library (https://docs.python. org/3/library /statistics.html) and then confirm your results match up to those that are
obtained when using the statistics module (where possible):




*   Mean
*   Median
*   Mode (hint: check out the Counter in the collections module of the standard library at https://docs.python. org/3/library /collections.html#collections. Counter)
*   Sample variance
*   Sample standard deviation




```python
# Mean
mean_calc = sum(salaries)/ len(salaries) #

print(f"Manual Mean: {mean_calc:,.2f}")
```

```python
# Median
# the middle value of the sorted lists

sorted_salaries = sorted(salaries)
n = len(sorted_salaries)
mid1 = sorted_salaries[n//2 - 1]
mid2 = sorted_salaries[n//2]
median_calc = (mid1 + mid2) / 2

print(f"Manual Median: {median_calc:,.2f}")
```

```python
# Mode

map_values = {}
for salary in salaries:
  if salary not in map_values:
    map_values[salary] = 1
    continue
    map_values[salary] += map_values[salary] + 1

mode_value = None
max_count = 0

for salary, count in map_values.items():
    if count > max_count:
        max_count = count
        mode_value = salary

print(f"Mode: {mode_value:,.2f} (appeared {max_count} times)")
```

```python
# Sample Variance

average = sum(salaries)/len(salaries)
len_minus_one = len(salaries) - 1

x_mean_squared = []
for salary in salaries:
  temp = (salary - average)**2
  x_mean_squared.append(temp)

  summation = sum(x_mean_squared)
  variance_calc = summation/len_minus_one

  print(f"Sample Variance: {variance_calc:,.2f}")
```

```python
# Sample Standard Deviation
# the square root of the variance
stdev_calc = variance_calc ** 0.5

print(f"Sample Standard Deviation: {stdev_calc:,.2f}")
```

Exercise 2

Using the same data, calculate the following statistics using the functions in the statistics module where appropriate:


*   Range
*   Coefficient of variation
*   Interquartile range
*   Quartile coefficient of dispersion


```python
# Range
import statistics

# the total spread of the data (Max minus Min)
salary_range = max(salaries) - min(salaries)

print(f"Range: {salary_range:,.2f}")
```

```python
# Coefficient of variation

# CV shows how large the standard deviation is relative to the mean (as a percentage)
mean_val = statistics.mean(salaries)
stdev_val = statistics.stdev(salaries)

coef_variation = (stdev_val / mean_val) * 100

print(f"Coefficient of Variation: {coef_variation:.2f}%")
```

```python
# Interquartile range

# IQR ignores outliers by measuring the spread of the middle 50% of the data
# divide the data into 4 quartiles to find the 25% (Q1) and 75% (Q3) marks
q1, q2, q3 = statistics.quantiles(salaries, n=4)
iqr = q3 - q1

print(f"Interquartile Range: {iqr:,.2f}")
```

```python
# Quartile coefficient of dispersion

# this uses the quartiles to measure spread; it's useful for comparing different datasets
# it is the ratio of the IQR (Q3-Q1) to the sum of the quartiles (Q3+Q1)
qcd = (q3 - q1) / (q3 + q1)

print(f"Quartile Coefficient of Dispersion: {qcd:.4f}")
```

```python
import pandas as pd
import numpy as np
df = pd.read_csv('diabetes.csv')

# 1. Identify column names
print("Column Names:", df.columns.tolist())

# 2. Identify data types
print("\nData Types:\n", df.dtypes)
```

```python
# 3. Total number of records
print(f"\nTotal Records: {len(df)}")
```

```python
# 4. Display first 20 records
# head() retrieves the top rows
print("\nFirst 20 Records:")
display(df.head(20))
```

```python
# 5. Display last 20 records
# tail() retrieves the bottom rows
print("\nLast 20 Records:")
display(df.tail(20))
```

```python
# 6. Change the Outcome column to Diagnosis
df = df.rename(columns={'Outcome': 'Diagnosis'})
```

```python
# 7. Create a new column Classification that display "Diabetes" if the value of outcome is 1 , otherwise "No Diabetes"
#1 becomes 'Diabetes', 0 becomes 'No Diabetes'
df['Classification'] = np.where(df['Diagnosis'] == 1, "Diabetes", "No Diabetes")

display(df[['Diagnosis', 'Classification']].head())
```

```python
# 8. Create a new dataframe "withDiabetes" that gathers data with diabetes
withDiabetes = df[df['Diagnosis'] == 1].copy() #diagnosis is 1

# 9. Create a new dataframe "noDiabetes" that gathers data with no diabetes
noDiabetes = df[df['Diagnosis'] == 0].copy()

# 10. Create a new dataframe "Pedia" that gathers data with age 0 to 19
# i used <= 19 to include everyone up to their 19th birthday
Pedia = df[df['Age'] <= 19].copy()

# 11. Create a new dataframe "Adult" that gathers data with age greater than 19
Adult = df[df['Age'] > 19].copy()

print("--- Subset Creation Successful ---")
print(f"Diabetes Count: {len(withDiabetes)}")
print(f"No Diabetes Count: {len(noDiabetes)}")
print(f"Pedia Count: {len(Pedia)}")
print(f"Adult Count: {len(Adult)}")
```

```python
# 12. Use numpy to get the average age and glucose value.
avg_age = np.mean(df['Age'])
avg_glucose = np.mean(df['Glucose'])

# 13. Use numpy to get the median age and glucose value
med_age = np.median(df['Age'])
med_glucose = np.median(df['Glucose'])

# 14. Use numpy to get the middle values (50th percentile)
mid_age = np.percentile(df['Age'], 50)
mid_glucose = np.percentile(df['Glucose'], 50)

# 15. Use numpy to get the standard deviation of the skinthickness
skin_std = np.std(df['SkinThickness'])

print(f"Average: Age={avg_age:.2f}, Glucose={avg_glucose:.2f}")
print(f"Median:  Age={med_age}, Glucose={med_glucose}")
print(f"Middle:  Age={mid_age}, Glucose={mid_glucose}")
print(f"SkinThickness Std Dev: {skin_std:.2f}")
```
